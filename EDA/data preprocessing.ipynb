```python
 X=data.drop(labels="Result",axis=1)
```

```python
Y=data["Result"]
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
wcss=[]
```

```python
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42) # initializing the KMeans object
    kmeans.fit(X) # fitting the data to the KMeans Algorithm
    wcss_iter = kmeans.inertia_ #within cluster sum of square 
    wcss.append(wcss_iter)

plot.plot(range(1,11),wcss) # creating the graph between WCSS and the number of clusters
plot.title('The Elbow Method')
plot.xlabel('Number of clusters')
plot.ylabel('WCSS')
plot.show()
```

```python
number_of_clusters = KneeLocator(range(1, 11), wcss, curve='convex', direction='decreasing')
```

```python
number_of_clusters.knee
```

```python
#Creating Clusters
```

```python
kmeans1 = KMeans(n_clusters=number_of_clusters.knee, init='k-means++', random_state=42)
```

```python
y_kmeans=kmeans1.fit_predict(X) #  divide data into clusters
```

```python
y_kmeans
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```
