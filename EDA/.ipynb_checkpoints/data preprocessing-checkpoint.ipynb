```python
 X=data.drop(labels="Result",axis=1)
```

```python
Y=data["Result"]
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
wcss=[]
```

```python
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42) # initializing the KMeans object
    kmeans.fit(X) # fitting the data to the KMeans Algorithm
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

plot.plot(range(1,11),wcss) # creating the graph between WCSS and the number of clusters
plot.title('The Elbow Method')
plot.xlabel('Number of clusters')
plot.ylabel('WCSS')
plot.show()
```

```python
number_of_clusters = KneeLocator(range(1, 11), wcss, curve='convex', direction='decreasing')
```

```python
number_of_clusters.knee
```

```python
#Creating Clusters
```

```python
kmeans1 = KMeans(n_clusters=number_of_clusters.knee, init='k-means++', random_state=42)
```

```python
y_kmeans=kmeans1.fit_predict(X) #  divide data into clusters
```

```python
y_kmeans
```

```python
X['Cluster']=y_kmeans 
```

```python
X
```

```python
X['Labels']=Y
```

```python
X
```

```python
for i in list_of_clusters:
    cluster_data=X[X['Cluster']==i] # filter the data for one cluster
     # Prepare the feature and Label columns
    cluster_features=cluster_data.drop(['Labels','Cluster'],axis=1)
    cluster_label= cluster_data['Labels']
    x_train, x_test, y_train, y_test = train_test_split(cluster_features, cluster_label, test_size=1 / 3, random_state=36)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics  import roc_auc_score,accuracy_score
```

```python
param_grid_xgboost = {

                "n_estimators": [100, 130], "criterion": ['gini', 'entropy'],
                               "max_depth": range(8, 10, 1)

            }
```

```python
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid_xgboost, verbose=3,cv=5)
```

```python
grid.fit(x_train, y_train)
```

```python
criterion = grid.best_params_['criterion']
max_depth = grid.best_params_['max_depth']
n_estimators = grid.best_params_['n_estimators']
```

```python
xgb = XGBClassifier(criterion=criterion, max_depth=max_depth,n_estimators= n_estimators, n_jobs=-1 )
           
```

```python
#Create Model for SVM
```

```python
param_grid = {"kernel": ['rbf', 'sigmoid'],
                          "C": [0.1, 0.5, 1.0],
                          "random_state": [0, 100, 200, 300]}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, cv=5,  verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
kernel = grid.best_params_['kernel']
C = grid.best_params_['C']
random_state = grid.best_params_['random_state']
```

```python
kernel
```

```python
sv_classifier = SVC(kernel=kernel,C=C,random_state=random_state)
```

```python
svm_model = sv_classifier.fit(x_train, y_train)
```

```python
prediction_svm=svm_model.predict(x_test)
```
